```python
data_copy["Size"] = data_copy["Size"].astype("float")
```

```python
data_copy["Size"].dtype
```

```python
# Since we have float values in our data sets which are like 4.6M. This became 4.6000. But in kb it should be 4.6 * 1000 = 4600
# Hence we need to multiply those values by 1000
```

```python
for i in data_copy["Size"]:
    if i < 10:
        data_copy["Size"]=data_copy["Size"].replace(i, i*1000)
```
